```python
# I will study tv data to answer questions like 
# What are the most extreme game outcomes 
# How does the game affect television vieweship 
# how have viewership, TV ratings and ad cost evolved 
# Who is the most prolific musician in terms of halftime viewership

# import modules 
import pandas as pd 

# load the data 
super_bowls = pd.read_csv('datasets/super_bowls.csv')
tv          = pd.read_csv('datasets/tv.csv')
halftime_musicians = pd.read_csv('datasets/halftime_musicians.csv')

# display the first five rows 
display(super_bowls.head(5))
display(tv.head(5))
display(halftime_musicians.head(5))
```

```python
# lets clean up the data  a little by first watching it 
from matplotlib import pyplot as plt 
# matplotlib inline 
plt.style.use('seaborn')

# plot a histogram of combined points 
super_bowls.combined_pts.hist()
plt.xlabel('combined points')
plt.ylabel('number of super bowls')
plt.show()

# display the super bowsl with highest and lowest combined scores 
display(super_bowls.query('combined_pts > 70'))
display(super_bowls.query('combined_pts < 25'))
```
